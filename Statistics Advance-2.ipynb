# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

## Probability Mass Function (PMF)
* The PMF is used for discrete random variables. These are variables that can take specific, separate values.

## Example: Rolling a Die

#### Imagine you have a fair six-sided die. When you roll the die, it can land on any one of six numbers: 1, 2, 3, 4, 5, or 6. Each number has an equal chance of showing up.

* The random variable X represents the result of the die roll.
* The PMF P(X=x) gives the probability that X equals a particular value X 
## Since the die is fair, the probability for each outcome is:

```python
P(X=1)=1/6,P(X=2)=1/6,P(X=3)=1/6,P(X=4)=1/6,P(X=5)=1/6,P(X=6)=1/6,
```

* In a bar graph, each outcome (1 through 6) would have a bar of height 1/6

## Probability Density Function (PDF)
* The PDF is used for continuous random variables. These are variables that can take any value within a range.

## Example: Heights of People

### Imagine measuring the heights of adult men in a country. Heights can vary continuously, like 170.2 cm, 175.6 cm, etc.

* The random variable Y represents a person's height.
* The PDF f(y) gives the relative likelihood of different heights.

```python
For simplicity, assume heights follow a normal distribution (bell curve) with an average (mean) height of 175 cm and a spread (standard deviation) of 10 cm.

The PDF for height Y looks like this:

f(y)=some formula with a bell shape

You can't use the PDF to find the exact probability of one height (like exactly 175 cm) because it's zero. Instead, you find the probability of a range of heights.

For example, to find the probability that a height is between 170 cm and 180 cm, you calculate the area under the PDF curve between 170 and 180.
```

# Key Differences
* PMF: Used for discrete outcomes (like rolling a die). It gives exact probabilities for each outcome.
* PDF: Used for continuous outcomes (like heights). It gives a relative likelihood and you find probabilities for ranges of outcomes.

## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

## Cumulative Density Function (CDF)
* The CDF gives you the probability that a random variable X is less than or equal to a certain value x.

## Example: Rolling a Die (Discrete Random Variable)
### Imagine rolling a fair six-sided die (like in a board game).

* Random Variable X: Represents the outcome of the die roll (values 1 through 6).

* CDF F(x): Tells us the probability that X is less than or equal to x.



## Why CDF is Useful
### Range Probability:
* You can find the probability that X falls within any range by subtracting F values. For example, P(2≤X≤4)=F(4)−F(1).

### Understanding Distribution:
* It helps in understanding how likely different outcomes are across the entire range of possible values X can take.

## Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.


## Normal Distribution
### Height of Individuals: 
* Heights of people in a large population tend to form a bell-shaped curve when plotted. Most people are near the average height, with fewer individuals being significantly shorter or taller.

### IQ Scores: 
* Intelligence Quotient (IQ) scores are often distributed in a bell curve pattern. The majority of people have average IQ scores, with fewer individuals having very high or very low scores.

### Measurement Errors: 
* Errors in measurements, like in experiments or instruments, often follow a bell curve distribution due to the central limit theorem. This theorem states that the distribution of many independent, random factors tends toward a normal distribution.

### Financial Markets: 
* Stock prices and returns over time can be modeled using a normal distribution, especially when considering large numbers of trades or data points due to statistical principles.

## Parameters of the Normal Distribution
### Mean (μ): 
* This is the average value around which data is centered. It determines the peak of the bell curve.

### Standard Deviation (σ): 
* This measures how spread out the data is around the mean. A larger standard deviation means data points are more spread out, creating a wider bell curve. A smaller standard deviation means data points are closer to the mean, resulting in a narrower bell curve.

## How Parameters Shape the Distribution
* Mean (μ): Moves the entire bell curve left or right along the x-axis. A higher mean shifts the curve to the right, and a lower mean shifts it to the left.

* Standard Deviation (σ): Controls the width of the bell curve. A larger standard deviation makes the curve wider and flatter, indicating more variability in the data. A smaller standard deviation makes the curve narrower and taller, showing less variability.

Practical Example
Imagine you're studying the heights of adult men:

* Mean (μ): Could be around 175 cm, representing the average height.
* Standard Deviation (σ): Might be 7 cm, showing how much variation exists around the average height.

## Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

## Bernoulli Distribution
* The Bernoulli distribution is used for a random variable that has exactly two possible outcomes: success (1) or failure (0). It is defined by a single parameter p, which is the probability of success.

## Example of Bernoulli Distribution
* Coin Toss: If you flip a fair coin, it can land on heads (success) or tails (failure).
* For a fair coin:
Probability of heads (p) = 0.5
Probability of tails (1−p) = 0.5

## Binomial Distribution
* The Binomial distribution is used for the number of successes in a fixed number of independent Bernoulli trials. It is defined by two parameters:
n: the number of trials
p: the probability of success in each trial

## Example of Binomial Distribution
* Multiple Coin Tosses: If you flip a fair coin 10 times: n=10 p=0.5
* The Binomial distribution tells you the probability of getting 0 to 10 heads in those 10 flips.

Key Differences
## Number of Trials:

* Bernoulli: One trial
* Binomial: Multiple trials (n)

## Outcomes:

* Bernoulli: Two possible outcomes (success or failure)
* Binomial: Number of successes in n trials

Example:

* Bernoulli: A single coin flip
* Binomial: Number of heads in 10 coin flips

### Summary
* Bernoulli Distribution: Models a single event with two outcomes.
* Binomial Distribution: Models multiple events, counting the number of succes

## Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

* To find the probability that a randomly selected observation from a normally distributed dataset with a mean of 50 and a standard deviation of 10 is greater than 60, we can use the standard normal distribution (Z-distribution).

## Steps to Calculate the Probability
* Standardize the observation: Convert the observation (60) to a Z-score using the formula:
Z= X−μ/σ

where:

* X is the value of the observation (60)
* μ is the mean (50)
* σ is the standard deviation (10)

* Look up the Z-score in the standard normal distribution table: This will give the probability that a value is less than the observed value (cumulative probability).

* Calculate the probability that the value is greater than the observed value: Subtract the cumulative probability from 1.


Conclusion
The probability that a randomly selected observation from this normally distributed dataset is greater than 60 is approximately 0.1587, or 15.87%.

## Q8: What is the z score? State the importance of the z score.

## What is the Z Score?
* A Z score measures how far a data point is from the mean of a dataset, in terms of standard deviations. It's calculated using the
# formula:

Z= X−μ/σ

where:

* X is the value of the data point.
* μ is the mean of the dataset.
* σ is the standard deviation of the dataset.

## Importance of the Z Score
### Standardization:

* Comparison Across Different Datasets: Z scores allow you to compare scores from different datasets by converting them to a common scale.
* Normalizing Data: Helps in analyzing data from different sources or scales by standardizing them.

### Probability and Percentiles:

* Finding Probabilities: Z scores help find the probability of a data point occurring within a normal distribution.
* Percentile Ranks: Indicates the relative standing of a data point within a dataset.

### Detection of Outliers:

* Identifying Outliers: Data points with Z scores greater than 3 or less than -3 are often considered outliers.

### Hypothesis Testing:

* Significance Testing: Z scores are used to determine if an observed result is statistically significant.
* Confidence Intervals: Helps calculate confidence intervals, showing the range within which the true population parameter likely falls.

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

## What is the Central Limit Theorem (CLT)?
* The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states:

"Given a sufficiently large sample size from a population with a finite mean and variance, the sampling distribution of the sample mean will be approximately normally distributed."

## Significance of the Central Limit Theorem
### Normal Distribution Approximation:

* Regardless of Population Distribution: Even if the population distribution is not normal, the sampling distribution of the sample mean tends to approach a normal distribution as the sample size increases. This is crucial because many statistical methods assume normality for inference.

## Foundation for Statistical Inference:

* Hypothesis Testing: The CLT forms the basis for many hypothesis tests, such as tests for means and proportions.
* Confidence Intervals: It allows the construction of confidence intervals around sample statistics, such as the mean or proportion.

## Practical Applications:

* Quality Control: It's used in quality control to analyze sample means and detect deviations from expected values.
* Survey Sampling: In survey sampling, it helps estimate population parameters from sample data.

## Sample Size Determination:

* Precision: It guides the determination of appropriate sample sizes to achieve desired levels of precision in estimation and hypothesis testing.
* Reliability: Ensures that statistical estimates are reliable and robust, particularly in studies where data variability can impact conclusions.

## Q10: State the assumptions of the Central Limit Theorem.

* Independent and Identically Distributed (iid) Samples: The samples you collect should be chosen randomly and should come from the same population, with each sample being independent of the others.

* Finite Mean and Variance: The population from which you're sampling should have a well-defined average (mean) and the values should not vary too widely from this average (variance should be finite).

* Sample Size: While there isn't a fixed minimum size, larger sample sizes generally lead to better results. Typically, a sample size of at least 30 is considered sufficient for the CLT to work well.

* No Extreme Outliers: The data should not have extreme values that significantly skew the results. Extreme outliers can affect the accuracy of the conclusions drawn using the CLT.
