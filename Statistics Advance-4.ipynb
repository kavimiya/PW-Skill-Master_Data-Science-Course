# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

## t-test:
* When to Use: Use this when you have a small sample size (less than 30) or don't know the population's standard deviation.
* Example: You have a group of 15 students, and you want to check if a new teaching method improves their test scores compared to their previous scores. Since you have a small group and don't know the standard deviation of all students' scores, you'd use a t-test.

## z-test:
* When to Use: Use this when you have a large sample size (30 or more) and you know the population's standard deviation.
* Example: A company knows the average and standard deviation of the time it takes to assemble a product. After introducing a new assembly method, they take a sample of 100 assembly times to see if the new method is faster. Since the sample is large and the standard deviation is known, they'd use a z-test.

# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error

# Type 1 Error (False Positive):
* Concept: Occurs when you reject a true null hypothesis. In other words, you conclude that there is an effect or difference when there actually isn’t.
* Example: Suppose a medical test incorrectly indicates that a patient has a disease (when they actually don’t). This is a Type 1 error—a false positive.
# Type 2 Error (False Negative):
* Concept: Occurs when you fail to reject a false null hypothesis. This means you conclude that there is no effect or difference when there actually is one.
* Example: Suppose a medical test fails to detect a disease that a patient actually has. This is a Type 2 error—a false negative

## Bayes's Theorem:

# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

# Margin of Error:
* The margin of error is how much your estimate might vary from the true value. It’s the range around your sample result where the true value is likely to be.

## Effect of Sample Size:
* Larger Sample Size: Makes the margin of error smaller.
* Smaller Sample Size: Makes the margin of error larger.
## Example Scenario:
* Suppose you want to know the average amount people spend on groceries.

* Small Sample Size (30 people): Margin of error might be $50. This means the true average spending could be $50 more or less than your sample average.

* Large Sample Size (300 people): Margin of error might be $10. With more people surveyed, your estimate is more precise, and the true average is likely within $10 of your sample average.

* Conclusion: A larger sample size gives a more accurate estimate with a smaller margin of error.

# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

## Hypothesis Test Using a t-Test
* 1. Hypotheses:
* Null Hypothesis  : The two teaching methods have the same effect on student performance.
* Alternative Hypothesis : The two teaching methods have different effects on student performance.
* 2. Sample Data:
* Sample A: Mean = 85, Standard Deviation = 6
* Sample B: Mean = 82, Standard Deviation = 5
* Assume both samples have 30 students each.

* 4. Critical t-Value:
* For a significance level of 0.01 with approximately 58 degrees of freedom, the critical t-value is about ±2.66.

* 5. Compare t-Statistic to Critical t-Value:
* t-Statistic: 2.10
* Critical t-Value: ±2.66
* Since 2.10 is less than 2.66, we do not reject the null hypothesis.

## Interpretation:
* There is not enough evidence to say that the two teaching methods result in significantly different student performance at the 0.01 significance level.
